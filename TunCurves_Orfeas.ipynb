$$ c_{pf} = \frac{1}{N_{spikes}} \sum_{i} L_i $$ 

where the sum on $i$ runs on all spikes emitted by a cell and $L_i$ is the location at which the i-th spike was emitted. Calculate this for each cell. Compare with the "true" place field centers from the simulations (which variables are we talking about?) 


```python
bins = np.arange(0., 200.)
track_length = 200. # the length of our linear track (eg in centimeter)
average_firing_rate = 5 # the peak firing rate, averaged across the population 
n_cells = 100 # how many cells we are recording
pf_centers = np.random.rand(n_cells) * track_length
pf_size = np.random.gamma(10, size=n_cells) # the size (width) of the place fields, drawn randomly from a gamma distribution 
pf_rate = np.random.exponential(scale=average_firing_rate, size=n_cells) 
n_runs= 20
av_running_speed = 10 # the average running speed (in cm/s)
fps = 30 # the number of "video frames" per second 
running_speed_a = np.random.chisquare(10, size=n_runs) # running speed in the two directions
running_speed_b = np.random.chisquare(10, size=n_runs) 

stopping_time_a = np.random.chisquare(15, size=n_runs) # the time the mouse will spend at the two ends of the track
stopping_time_b = np.random.chisquare(15, size=n_runs)

x = np.array([])

for i in range(n_runs):
    stop1 = np.ones((int(stopping_time_a[i]*fps),)) * 0.
    run_length = len(bins) * fps / running_speed_a[i]
    run1 = np.linspace(0., float(len(bins)-1), int(run_length))
    stop2 = np.ones((int(stopping_time_b[i]*fps),)) * (len(bins)-1.)
    run_length = len(bins) * fps / running_speed_b[i]
    run2 = np.linspace(len(bins)-1., 0., int(run_length))
    x = np.concatenate((x, stop1, run1, stop2, run2))
t = np.arange(len(x))/fps
```

```python
true_firing_rate_maps = np.zeros((n_cells, len(bins)))
for i in range(n_cells):
    true_firing_rate_maps[i,:] = pf_rate[i] * np.exp(-((bins-pf_centers[i])**2)/(2*pf_size[i]**2))
sampling_rate = 10000.

t_sampling = np.arange(0, t[-1], 1. / sampling_rate)
x_sampling = np.floor(np.interp(t_sampling, t, x))
noise_firing_rate = 0.1 # the baseline noise firing rate
spikes = []

for i in tqdm(range(n_cells)):
    inst_rate = true_firing_rate_maps[i,x_sampling.astype(np.int32)] + noise_firing_rate
    spikes_loc = np.random.poisson(inst_rate/sampling_rate)
    sp = np.argwhere(spikes_loc)
    t_sp = t_sampling[sp]
    spikes.append(t_sp)
```

```python
space_bins = np.arange(0., 200., 5.)
coarse=np.arange(50., 100., 5.)
broad=np.arange(0., 400., 5.)
spike_positions = [np.interp(s, t, x) for s in spikes]

# we compute histograms for each cell
spikes_hist= [np.histogram(s, space_bins)[0] for s in spike_positions]
coa=[np.histogram(s, coarse)[0] for s in spike_positions]
bro=[np.histogram(s, broad)[0] for s in spike_positions]

# put them together into a matrix of floating point numbers (for plotting)
spikes_hist = np.vstack(spikes_hist).astype(np.float64)
coa=np.vstack(coa).astype(np.float64)
bro=np.vstack(bro).astype(np.float64)

# we also need an "occupancy histogram" in order to normalize the firing rates maps 
occupancy = np.histogram(x, space_bins)[0] /  fps
occupancyc = np.histogram(x, coarse)[0] /  fps
occupancyb = np.histogram(x, broad)[0] /  fps


default = spikes_hist / occupancy 
coar=coa/occupancyc
broa=bro/occupancyb
```

```python
fig=plt.figure(figsize=(10,5))
sub1=fig.add_subplot(2,2,1)
sub2=fig.add_subplot(2,2,2)
sub3=fig.add_subplot(2,2,3)
sub1.title.set_text('Default')
sub2.title.set_text('Coarse')
sub3.title.set_text('Broad')
plt.subplots_adjust(hspace=0.5,wspace=0.1)



sub1.imshow(default, cmap='inferno', extent = [0, 200, 0, n_cells])
sub2.imshow(coar, cmap='inferno', extent = [0, 200, 0, n_cells])
sub3.imshow(broa, cmap='inferno', extent = [0, 200, 0, n_cells])
```

```python
lown=0.01
highn=1
spikesln=[]
spikeshn=[]

for i in tqdm(range(n_cells)):
    inst_rate = true_firing_rate_maps[i,x_sampling.astype(np.int32)] + lown
    spikes_loc = np.random.poisson(inst_rate/sampling_rate)
    sp = np.argwhere(spikes_loc)
    t_spln = t_sampling[sp]
    spikesln.append(t_spln)
    

for i in tqdm(range(n_cells)):
    inst_rate = true_firing_rate_maps[i,x_sampling.astype(np.int32)] + highn
    spikes_loc = np.random.poisson(inst_rate/sampling_rate)
    sp = np.argwhere(spikes_loc)
    t_sphn = t_sampling[sp]
    spikeshn.append(t_sphn)
    
```

```python
space_bins = np.arange(0., 200., 5.)
spikesln_positions = [np.interp(s, t, x) for s in spikesln]
spikeshn_positions = [np.interp(s, t, x) for s in spikeshn]


# we compute histograms for each cell
spikesln_hist= [np.histogram(s, space_bins)[0] for s in spikesln_positions]
spikeshn_hist= [np.histogram(s, space_bins)[0] for s in spikeshn_positions]



# put them together into a matrix of floating point numbers (for plotting)
spikesln_hist = np.vstack(spikesln_hist).astype(np.float64)
spikeshn_hist = np.vstack(spikeshn_hist).astype(np.float64)


# we also need an "occupancy histogram" in order to normalize the firing rates maps 
occupancy = np.histogram(x, space_bins)[0] /  fps



lownoise = spikesln_hist / occupancy 
highnoise=spikeshn_hist/occupancy
```

```python
fig2=plt.figure(figsize=(10,5))
sub11=fig2.add_subplot(2,2,1)
sub22=fig2.add_subplot(2,2,2)
plt.subplots_adjust(hspace=0.5,wspace=0.1)

sub11.title.set_text('Low Noise (0.01)')
sub22.title.set_text('High Noise (1)')

sub11.imshow(lownoise, cmap='inferno', extent = [0, 200, 0, n_cells])
sub22.imshow(highnoise, cmap='inferno', extent = [0, 200, 0, n_cells])
```
